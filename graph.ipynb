```
import numpy as np
import matplotlib.pyplot as plt
```

```
import random

x = np.linspace(0,60,60)
noise = np.random.normal(0,5,100)
noise1 = np.random.normal(0,0.2,100)

sb = [32,34,35,36, 40,41,42]
gadha = [19,20,21, 57,58]
b_gadha = [47,48,49,50]
y = 5*np.sin(x)

for i in range(60):
  if i in sb:
    y[i] = y[i] + noise[i]
  if i in gadha :
    y[i] = y[i] + 0.4*noise[i]*random.randrange(10)
  if i in b_gadha :
    y[i] = y[i] + 0.7*noise[i]
  else:
    y[i] = y[i] + abs(noise1[i])
plt.plot(x,y, '-g', label='sin(x)')
```

```
const_line = [0 for x in range(1, 101)]
noise = np.random.normal(0, 5, 100)
noise1 = np.random.normal(0, 0.2, 100)

sb = [32,33,34,35, 40,41,42]
pothole = [19,20,21, 58,59]
big_pothole = [47,48,49,50]

for i in range(100):
  if i in sb:
    val = random.uniform(-10.0, 10.0)
    while(val<5.0 and val>-5.0):
      val = random.uniform(-10.0, 10.0)
    const_line[i] = const_line[i] + val
  elif i in pothole:
    const_line[i] = const_line[i] + np.clip(0.4*noise[i], -5, 5)
  elif i in big_pothole:
    const_line[i] = const_line[i] + 0.6*noise[i]
  else:
    const_line[i] = const_line[i] + 0.03*noise[i]

plt.figure(figsize=(15, 15))
plt.plot([x for x in range(1, 101)], const_line, '-gD', markevery = pothole)
plt.plot([x for x in range(1, 101)], const_line, '-rD', markevery = sb)
plt.plot([x for x in range(1, 101)], const_line, '-yD', markevery = big_pothole)
```

```
const_line = np.array(const_line)
np.save('./car_data.npy', const_line)
```

```
timestep = [i for i in range(1, 101)]

for i in timestep:
  if i in sb:
    plt.figure(figsize=(10, 10))
    plt.plot([x for x in range(1, 101)], const_line, '-rD', markevery = [i], markersize=8)
    plt.savefig("./images/"+str(i)+'.png')
  elif i in pothole:
    plt.figure(figsize=(10, 10))
    plt.plot([x for x in range(1, 101)], const_line, '-yD', markevery = [i], markersize=8)
    plt.savefig("./images/"+str(i)+'.png')
  elif i in big_pothole:
    plt.figure(figsize=(10, 10))
    plt.plot([x for x in range(1, 101)], const_line, '-gD', markevery = [i], markersize=8)
    plt.savefig("./images/"+str(i)+'.png')
  else:
    plt.figure(figsize=(10, 10))
    plt.plot([x for x in range(1, 101)], const_line, '-bD', markevery = [i], markersize=8)
    plt.savefig("./images/"+str(i)+'.png')
```

```
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib import style

style.use('fivethirtyeight')

fig = plt.figure()
ax1 = fig.add_subplot(1,1,1)
```

```
def animate(i):
    graph_data = open('example.txt','r').read()
    lines = graph_data.split('\n')
    xs = []
    ys = []
    for line in lines:
        if len(line) > 1:
            x, y = line.split(',')
            xs.append(float(x))
            ys.append(float(y))
    ax1.clear()
    ax1.plot(xs, ys)
```

```
ani = animation.FuncAnimation(fig, animate, interval=1000)
plt.show()
```
