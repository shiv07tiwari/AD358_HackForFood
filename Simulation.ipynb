```
import numpy as np
import matplotlib.pyplot as plt
```

```
import random

x = np.linspace(0,60,60)
noise = np.random.normal(0,5,100)
noise1 = np.random.normal(0,0.2,100)

sb = [32,33,34,35, 40,41,42]
gadha = [19,20,21, 58,59]
b_gadha = [47,48,49,50]
y = 5*np.sin(x)

for i in range(60):
  if i in sb:
    y[i] = y[i] + noise[i]
  if i in gadha :
    y[i] = y[i] + 0.4*noise[i]*random.randrange(10)
  if i in b_gadha :
    y[i] = y[i] + 0.7*noise[i]
  else:
    y[i] = y[i] + abs(noise1[i])
plt.plot(x,y, '-g', label='sin(x)')
```

```
const_line = [0 for x in range(1, 101)]
noise = np.random.normal(0, 5, 100)
noise1 = np.random.normal(0, 0.2, 100)

sb = [32,33,34,35, 40,41,42]
pothole = [19,20,21, 58,59]
big_pothole = [47,48,49,50]

for i in range(100):
  if i in sb:
    const_line[i] = const_line[i] + np.clip(1.0*noise[i], -10.0, 10.0)
  elif i in pothole:
    const_line[i] = const_line[i] + np.clip(0.4*noise[i], -5.0, 5.0)
  elif i in big_pothole:
    const_line[i] = const_line[i] + 0.9*noise[i]
  else:
    const_line[i] = const_line[i] + 0.05*noise[i]

plt.figure(figsize=(15, 15))
plt.plot([x for x in range(1, 101)], const_line, '-gD', markevery = pothole)
plt.plot([x for x in range(1, 101)], const_line, '-rD', markevery = sb)
plt.plot([x for x in range(1, 101)], const_line, '-yD', markevery = big_pothole)
```
